```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Paths to the dataset
dataset_path = r'C:\Users\blneh\Documents\Tomato1'  # Adjust this path to your dataset

# Define ImageDataGenerators for 70:30 train-validation split
datagen = ImageDataGenerator(
    rescale=1./255,
    horizontal_flip=True,
    rotation_range=20,
    validation_split=0.3  # 70% training, 30% validation
)

train_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical',  # Use categorical for multi-class classification
    subset='training'
)

val_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical',  # Use categorical for multi-class classification
    subset='validation'
)

# Build the MobileNet model
base_model = MobileNet(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(3, activation='softmax')(x)  # 3 classes and softmax activation

model = Model(inputs=base_model.input, outputs=predictions)

# Freeze all layers in MobileNet except the last few for fine-tuning
for layer in base_model.layers[:-20]:
    layer.trainable = False

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    train_generator,
    epochs=100,  # Adjust epochs as needed
    validation_data=val_generator,
    verbose=1
)

# Evaluate the model
val_steps = int(np.ceil(val_generator.samples / val_generator.batch_size))
y_pred_prob = model.predict(val_generator, steps=val_steps)
y_pred = np.argmax(y_pred_prob, axis=1)
y_true = val_generator.classes

# Compute metrics
conf_matrix = confusion_matrix(y_true, y_pred)
f1 = f1_score(y_true, y_pred, average='macro')
precision = precision_score(y_true, y_pred, average='macro')
sensitivity = recall_score(y_true, y_pred, average='macro')

# Accuracy, loss from training history
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

# Print metrics
print(f'Final Accuracy: {acc[-1]:.4f}')
print(f'Final Validation Accuracy: {val_acc[-1]:.4f}')
print(f'Final Loss: {loss[-1]:.4f}')
print(f'Final Validation Loss: {val_loss[-1]:.4f}')
print(f'F1 Score: {f1:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Sensitivity (Recall): {sensitivity:.4f}')

# Plot training & validation accuracy and loss
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(acc, label='Train Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(loss, label='Train Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=val_generator.class_indices.keys(),
            yticklabels=val_generator.class_indices.keys())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.show()
```
